# Exploratory Data Analysis

```python
#Which brand of phones are popular for gaming?
plt.figure(figsize=(10, 6))
sns.countplot(data=phone_data, y='phone brand', order=phone_data['phone brand'].value_counts().index, hue='phone brand', palette='viridis', legend=False)
plt.title('Popularity of Phone Brands')
plt.xlabel('Count')
plt.ylabel('Phone Brand')
plt.show()
```

```python
#What is the primary use of phone observed in the dataset?
plt.figure(figsize=(8, 8))
sizes = phone_data['primary use'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', colors=['#ffcc00', '#66b3ff', '#99ff99', '#ff6666'], startangle=90, wedgeprops={'edgecolor': 'black', 'width': 0.3})
plt.title('Primary Use of Phone (Donut Chart)')
plt.ylabel('')  # To hide the ylabel
plt.show()
```

```python
#What is the distribution of monthly recharge costs in the dataset?
plt.figure(figsize=(10, 6))
sns.histplot(data=phone_data, x='monthly recharge cost (inr)', kde=True, color='purple', bins=15)
plt.title('Monthly Recharge Cost Distribution')
plt.xlabel('Recharge Cost (₹)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Which brands of phones are highly used for Gaming?
gaming_data = phone_data[phone_data['primary use'] == 'Gaming']

# Get count of phone brands used for gaming
gaming_brand_counts = gaming_data['phone brand'].value_counts()

# Create the area chart
plt.figure(figsize=(10, 6))
plt.stackplot(gaming_brand_counts.index, gaming_brand_counts.values, labels=gaming_brand_counts.index, alpha=0.6, color=plt.cm.coolwarm(range(len(gaming_brand_counts))))

# Customize the chart
plt.title('Phone Brands Most Used for Gaming (Area Chart)')
plt.xlabel('Phone Brand')
plt.ylabel('Count')
plt.legend(title='Phone Brands', loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#How does data usage vary across different age groups?
plt.figure(figsize=(10, 6))
sns.lineplot(data=phone_data, x='age', y='data usage (gb/month)', marker='o', color='orange')
plt.title('Trend of Data Usage with Age')
plt.xlabel('Age')
plt.ylabel('Data Usage (GB/month)')
plt.show()
```

```python
#What is the Ecommerce spending distribution according to age?
# Grouping by 'age' and calculating the mean e-commerce spend (inr/month)
age_ecommerce_line = phone_data.groupby('age')['e-commerce spend (inr/month)'].mean().reset_index()

# Sorting the data by age
age_ecommerce_line = age_ecommerce_line.sort_values('age')

# Plotting the line plot
plt.figure(figsize=(12, 6))
plt.plot(age_ecommerce_line['age'], 
         age_ecommerce_line['e-commerce spend (inr/month)'], 
         marker='o', color='teal', linestyle='-', linewidth=2, markersize=6)

# Adding titles and labels
plt.title('E-Commerce Spend vs Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average E-Commerce Spend (INR)', fontsize=14)
plt.grid(True)

# Display the plot
plt.show()
```

```python
#What is the relationship between social media usage time and ecommerce usage time, and how does data usage impact this relationship?
plt.figure(figsize=(10, 6))
plt.scatter(
    phone_data['social media time (hrs/day)'], 
    phone_data['e-commerce spend (inr/month)'], 
    s=phone_data['data usage (gb/month)'] * 10,  # Bubble size
    alpha=0.6, c=phone_data['data usage (gb/month)'], cmap='viridis', edgecolor='k'
)
plt.colorbar(label='Data Usage (GB/month)')
plt.title('Bubble Chart: Social Media Time vs Ecommerce Time', fontsize=16)
plt.xlabel('Social Media Time (in minutes)')
plt.ylabel('Ecommerce Time (in minutes)')
plt.show()
```

```python
#How is screen time correlated with other usage metrics like data usage, call duration, and recharge cost?
plt.figure(figsize=(10, 6))
correlation_matrix = phone_data[['screen time (hrs/day)', 'data usage (gb/month)', 'calls duration (mins/day)', 'monthly recharge cost (inr)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Heatmap: Screen Time vs Other Variables')
plt.show()
```

```python
#What is the gender distribution across different phone brands?
# Create a count of gender and phone brand combinations
gender_phone_counts = phone_data.groupby(['phone brand', 'gender']).size().reset_index(name='count')

# Pivot the data to prepare for stacked bar plotting
pivot_table = gender_phone_counts.pivot(index='phone brand', columns='gender', values='count').fillna(0)

# Plotting the stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#1f77b4', '#ff7f0e'], edgecolor='black')

# Customize the chart
plt.title('Gender vs Phone Brand', fontsize=16)
plt.xlabel('Phone Brand', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# What is the preferred operating system across different age groups?
plt.figure(figsize=(12, 6))
sns.countplot(data=phone_data, x='age', hue='os', palette='Set1')

# Customizing the plot
plt.title('Age vs Operating System (OS)', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Operating System')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
